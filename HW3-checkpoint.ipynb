```python
# Often when we process a dataset, we use multiple different data sources for finer 
# analysis of our data. In this case, the dataset, contains district number however it would make 
# easier to analysis data based on district if we can also know, what name each district number represents.
# For eg: district number 25 represents Austin district in Chicago.

# To answer such question, I digged further into the source of the dataset, and found that this link:
# https://data.cityofchicago.org/Public-Safety/Boundaries-Police-Districts-current-/fthy-xz3r contains the 
# various file format to download (Click on Export) the information of the datasets. 
# For eg: geoJSON file format contains the "area_numbe" value and "district" which contains the name

# Simiarly, The following code reads the geojson file (other file format than csv) and then create a 
# community name based on the community number


import json

community_df = pd.DataFrame()
with open('datasets/Boundaries - Community Areas (current).geojson') as f:
    js = json.load(f)
    
    for feature in js['features']:
        row = {}
        num = feature['properties']['area_numbe']
        name = feature['properties']['community']
        row['Community Area'] = num
        row['community'] = name
        community_df = community_df.append(row, ignore_index=True)
```

```python
#Question 16: Convert the Community Area columns type from float64 to object (string) in the crimes_df. 
# Note to remove the decimal portion from the dataset
crimes_df['Community Area'] = crimes_df['Community Area'].apply(int)
crimes_df['Community Area'] = crimes_df['Community Area'].astype(str)
```

```python
# Question 17: Merge the community_df with crimes_df (merge using intersection), such that there will be one additional column ('Community Name') 
# in the crimes_df with the name of the Communty Area number.
crimes_df.reset_index()
isinstance(crimes_df['Community Area'].iloc[1], type(community_df['community'].iloc[1]))
crimes_df = crimes_df.merge(community_df, on='Community Area')
crimes_df = crimes_df.rename(columns={'community': 'Community Name'})
crimes_df.head()
# Hint: please make sure both the keys are same type before you merge them together
# Also, first reset the index before merging
```
